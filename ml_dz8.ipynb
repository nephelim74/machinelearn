Решите задачу классификации оттоковых клиентов телекома: для каждого клиента необходимо определить, перестанет он пользоваться услугами оператора в ближайшее время или нет.

Сделайте первичную обработку данных:

1. Выбросите колонку State.

2. Колонки International plan и Voice mail plan закодируйте следующим образом: Yes замените на 1, а No на 0.

3. Переведите целевую переменную (Churn) в числа: True в 1, а False в 0.

```python
import pandas as pd

# Загружаем данные
url = "https://gbcdn.mrgcdn.ru/uploads/asset/6057170/attachment/326b1a69e5645a4a2f66fdad2177bac7.csv"
data = pd.read_csv(url)
data.head()
```

```python

# Удаляем колонку State
data = data.drop("State", axis=1)

# Кодируем колонки International plan и Voice mail plan
data["International plan"] = data["International plan"].map({"Yes": 1, "No": 0})
data["Voice mail plan"] = data["Voice mail plan"].map({"Yes": 1, "No": 0})

# Переводим целевую переменную Churn в числа
data["Churn"] = data["Churn"].astype(int)

# Выводим первые строки для проверки
print(data.head())
```

Выведите на экран первые строки изменённой таблицы, чтобы проверить, что все преобразования сработали.

```python
data.head()
```

Создайте матрицу объект-признак X и вектор с целевой переменной (Churn) y.

```python
# ваш код здесь
# Создаем матрицу объект-признак X и вектор целевой переменной y
X = data.drop("Churn", axis=1)
y = data["Churn"]
```

Проверьте баланс классов. Является ли выборка сбалансированной?

```python
# ваш код здесь
# Проверяем баланс классов
print(y.value_counts())
```

Разбейте данные на тренировочную и тестовую части, размер тестовой части — 30%.

```python
from sklearn.model_selection import train_test_split

# ваш код здесь
from sklearn.model_selection import train_test_split

# Разделяем данные на тренировочную и тестовую части (70% - тренировка, 30% - тест)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Для линейных моделей очень важно масштабирование признаков. Поэтому отмасштабируйте признаки при помощи [StandardScaler](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html), обучив метод на тренировочных данных и применив его и к трейну, и к тесту.

```python
from sklearn.preprocessing import StandardScaler

# ваш код здесь
# Масштабируем признаки
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

```

После масштабирования признаков X_train и X_test перестали быть объектами типа pd.DataFrame, это неудобно. Приведите их снова к pd.DataFrame.

```python
# ваш код здесь
# Преобразуем обратно в DataFrame для удобства
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X.columns)
```

Обучите на тренировочных данных с параметрами по умолчанию:

1) логистическую регрессию;

2) метод опорных векторов (с kernel='linear').

```python
from sklearn.linear_model import LogisticRegression

# Обучаем модель
log_reg = LogisticRegression()
log_reg.fit(X_train_scaled, y_train)

# Предсказание на тесте
y_pred_log_reg = log_reg.predict(X_test_scaled)
```

```python
from sklearn.svm import SVC

# Обучаем модель
svm = SVC(kernel='linear', probability=True)
svm.fit(X_train_scaled, y_train)

# Предсказание на тесте
y_pred_svm = svm.predict(X_test_scaled)
```

С помощью обученных моделей сделайте предсказание на тестовых данных.

```python
from sklearn.metrics import roc_auc_score

# ваш код здесь
from sklearn.metrics import roc_auc_score

# Оценка для логистической регрессии
roc_auc_log_reg = roc_auc_score(y_test, log_reg.predict_proba(X_test_scaled)[:, 1])
print("ROC-AUC для логистической регрессии:", roc_auc_log_reg)

# Оценка для SVM
roc_auc_svm = roc_auc_score(y_test, svm.predict_proba(X_test_scaled)[:, 1])
print("ROC-AUC для SVM:", roc_auc_svm)
```

Теперь обучите решающее дерево с параметрами по умолчанию на тренировочных данных, сделайте предсказание на тесте. Используйте метрику для оценки.

```python
from sklearn.tree import DecisionTreeClassifier

# ваш код здесь
from sklearn.tree import DecisionTreeClassifier

# Обучаем решающее дерево
tree = DecisionTreeClassifier(random_state=42)
tree.fit(X_train, y_train)

# Предсказание на тесте
y_pred_tree = tree.predict(X_test)

# Оценка
roc_auc_tree = roc_auc_score(y_test, tree.predict_proba(X_test)[:, 1])
print("ROC-AUC для решающего дерева:", roc_auc_tree)
```

```python
# ваш код здесь
# Разделяем данные заново без масштабирования
X_train_new, X_test_new, y_train_new, y_test_new = train_test_split(X, y, test_size=0.3, random_state=42)

# Обучаем решающее дерево без масштабирования
tree_new = DecisionTreeClassifier(random_state=42)
tree_new.fit(X_train_new, y_train_new)

# Предсказание и оценка
y_pred_tree_new = tree_new.predict(X_test_new)
roc_auc_tree_new = roc_auc_score(y_test_new, tree_new.predict_proba(X_test_new)[:, 1])
print("ROC-AUC для решающего дерева без масштабирования:", roc_auc_tree_new)
```

Какая модель справилась с решением задачи?
