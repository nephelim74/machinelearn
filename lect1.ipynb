```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
y = np.array([0, 1, 2, 3, 4, 5, 6])
```

```python
plt.scatter(x, y, s=30)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
y1 = np.array([0, 1, 2, 3, 4, 5, 6])
y2 = np.array([0, 1, 2, 10, 4, 3, 8])
y3 = np.array([0, 1, 3, 1, 5, 4, 3])
plt.scatter(x, y, s=30)
plt.plot(y1, 'g')
plt.plot(y2, 'r')
plt.plot(y3, 'b')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
e1 = sum(y - y1)
e2 = sum(y - y2)
e3 = sum(y - y3)
print(e1)
print(e2)
print(e3)
```

```python
e1 = sum(y - y1)/len(y)
e2 = sum(y - y2)/len(y)
e3 = sum(y - y3)/len(y)
print(e1)
print(e2)
print(e3)
```

```python
e1 = sum(abs(y - y1))/len(y)
e2 = sum(abs(y - y2))/len(y)
e3 = sum(abs(y - y3))/len(y)
print(e1)
print(e2)
print(e3)
```

```python
from sklearn.metrics import mean_absolute_error
e1 = mean_absolute_error(y, y1)
e2 = mean_absolute_error(y, y2)
e3 = mean_absolute_error(y, y3)
print(e1)
print(e2)
print(e3)
```
